```python
import requests; from urllib.parse import urlparse
import pandas as pd
```

```python
df = pd.read_csv('subway.csv')
df
```

```python
#카카오 api 를 이용해 도로명 주소를 x,y 경도 위도로 변환
def coordinate(address) :
    url = "https://dapi.kakao.com/v2/local/search/address.json?&query=" +address
    result = requests.get(urlparse(url).geturl(), headers={"Authorization": "KakaoAK {인증키}"})
    json_obj = result.json()
    val = []
    for document in json_obj['documents']:
        val = [document['address_name'], document['x'], document['y']]
        
    return val
```

```python
#도로명 주소 하나만 변환할때
coordinate('대전 중구 어덕마을로 1-6') 
```

```python
#파일로된 도로명 주소들의 x,y 좌표 값을 구함
list = []
for addres in df['name']:
    list.append(coordinate(addres))
df1 = pd.DataFrame(list, columns = ['address_name', 'x', 'y'])
df1.head()
```

```python
#변환된 결과물 파일로 출력
df1.to_excel('C:/Users/user/대전 지하철.xlsx')
```
