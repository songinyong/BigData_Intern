# 대전시 유동인구 데이터 구별로 정제할때 코드

```python
import pandas as pd
```

```python
western_50 = pd.read_csv('서구50격자.csv')
```

```python
western_250 = pd.read_csv('서구250격자.csv')
```

```python
western_50.head(3)
```

```python
western_250.head(3)
```

```python
western = pd.merge(western_250,western_50, left_on='gid', right_on='cell_250_id', how='left')
```

```python
western.head(3)
```

```python
#data = pd.read_csv('DAEJEON_20191001_20191015.csv')#10월
#data = pd.read_csv('DAEJEON_20191101_20191115.csv')#11월
#data = pd.read_csv('DAEJEON_20191201_20191215.csv')#12월
#data = pd.read_csv('DAEJEON_20200301_20200321.csv')#20년 3월
data = pd.read_csv('DAEJEON_20200401_20200415.csv')#20년 4월
data.head()
```

```python
#data2 = pd.read_csv('DAEJEON_20191016_20191031.csv')#10월
#data2 = pd.read_csv('DAEJEON_20191116_20191130.csv')#11월
#data2 = pd.read_csv('DAEJEON_20191216_20191231.csv')#12월
#data2 = pd.read_csv('DAEJEON_20200322_20200331.csv')#20년 3월
data2 = pd.read_csv('DAEJEON_20200416_20200430.csv')#20년 4월
data2.head()
```

```python
#분석을 위해 하나의 파일로 합침
month_data = pd.concat([data, data2])
```

```python
#달 데이터가 통합되어 있을때
#month_data = pd.read_csv('DAEJEON_20200101_20200131.csv')#20년 1월
#month_data = pd.read_csv('DAEJEON_20200201_20200229.csv')#20년 2월
#month_data = pd.read_csv('DAEJEON_20200501_20200531.csv')#20년 5월
month_data = pd.read_csv('DAEJEON_202006.csv')#20년 6월
```

```python
month_data.head(1)
```

```python
#필요없는 속성값만 지정해서 제거
del month_data['x']
del month_data['y']
del month_data['etl_ymd']
del month_data['timezn_cd']
```

```python
#피봇 테이블 이용해 각 id별 유동인구 데이터 합침
pdf1 = pd.pivot_table(month_data,                
                     index = 'id',          
                     aggfunc = 'sum')
```

```python
#조인해서 대전시 전체 유동인구 격자 데이터중 서구에 속한 격자 데이터만 가져옴
western_join = pd.merge(western,pdf1, left_on='cell_50_id', right_on='id', how='left')
```

```python
#결측값은 유동인구 측정이 안되는 부분이니 0으로 처리 
western_join = western_join.fillna(0)
western_join.head()
```

```python
western_join.columns
```

```python
#일 평균값으로 바꾼후 시간별 평균으로 나눔
western_join.iloc[:, 15:] = western_join.iloc[:, 15:]/30 #그 달의 월 값입력

western_join.iloc[:, 15:] = western_join.iloc[:, 15:]/24 #시간이니 고정
```

```python
del western_join['GU']
del western_join['cell_50_id']
del western_join['cell_250_id']
del western_join['cell_500_id']
del western_join['x_katech']
del western_join['y_katech']
del western_join['city_cd']
del western_join['city_nm']
del western_join['gu_cd']
del western_join['gu_nm']
del western_join['hj_dong_cd']
del western_join['hj_dong_nm']
del western_join['bj_dong_cd']
del western_join['bj_dong_nm']
```

```python
pdf2 =pd.pivot_table(western_join,                
                     index = 'gid',          
                     aggfunc = 'sum')
```
