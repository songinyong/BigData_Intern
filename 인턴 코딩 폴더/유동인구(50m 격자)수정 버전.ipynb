```python
#분석을 위해 하나의 파일로 합침
month_data = pd.concat([data, data2])
```

```python
#필요없는 속성값만 지정해서 제거
del month_data['x']
del month_data['y']
del month_data['etl_ymd']
del month_data['timezn_cd']
```

```python
#피봇 테이블 이용해 각 id별 유동인구 데이터 합침
pdf1 = pd.pivot_table(month_data,                
                     index = 'id',          
                     aggfunc = 'sum')
```

```python
#조인해서 대전시 각 격자에 값 저장
daejeaon_join = pd.merge(daejeaon,pdf1, left_on='cell_50_id', right_on='id', how='left')
```

```python
#결측값은 유동인구 측정이 안되는 부분이니 0으로 처리 
daejeaon_join = daejeaon_join.fillna(0)
daejeaon_join.head()
```

```python
pdf2 = pd.pivot_table(daejeaon_join,                
                     index = 'cell_250_id',          
                     aggfunc = 'sum')
```

```python
pdf2
```

```python
#조인해서 대전시 전체 250m 격자 데이터에 값 저장
daejeaon250_join = pd.merge(daejeaon_250,pdf2, left_on='gid', right_on='cell_250_id', how='left')
```

```python
del daejeaon250_join['bj_dong_cd']
del daejeaon250_join['cell_50_id']
del daejeaon250_join['city_cd']
del daejeaon250_join['x_katech']
del daejeaon250_join['y_katech']
del daejeaon250_join['gu_cd']
del daejeaon250_join['hj_dong_cd']
```

```python
#null 값은 전부 0으로 처리
daejeaon250_join = daejeaon250_join.fillna(0)
```

```python
daejeaon250_join.to_excel('유동인구1차정제/20년{}월 대전시 유동인구(250).xlsx'.format(month))
```
